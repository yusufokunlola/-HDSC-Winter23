# Tasks

From the dataset, fit a linear model on the relationship between the temperature in the living room in Celsius (x = T2) and the temperature outside the building (y = T6). 

What is the R^2 value in two d.p.?

```python
# instantiate response and target variables
X = df[["T2"]]
y = df[["T6"]]

# Split data into feature matrix `X_train` and target vector `y_train`.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)


# normalize data using MinMaxScaler
ss = MinMaxScaler()
scalled_x_train = ss.fit_transform(X_train)
scalled_x_test = ss.transform(X_test)

# fit model to scaled data
model = LinearRegression().fit(scalled_x_train, y_train)

# predict on scaled test data
y_pred = model.predict(scalled_x_test)

# calculate r2 score
score = r2_score(y_test, y_pred)
print('R-squared:', round(score, 2)) 
```

Normalize the dataset using the MinMaxScaler after removing the following columns: [“date”, “lights”]. The target variable is “Appliances”. Use a 70-30 train-test set split with a random state of 42 (for reproducibility). Run a multiple linear regression using the training set and evaluate your model on the test set. 

Answer the following questions:

What is the Mean Absolute Error (in two decimal places)?

### **Preparing the data**

```python
# drop date and lights columns
df.drop(columns=["date", "lights"], inplace=True)
```

```python
# instantiate a minmax scaler for normalisation

scaler = MinMaxScaler()
normalised_df = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
```

```python
# The next step is to divide the data into "attributes" (inputs) and "labels" (outputs)
target = "Appliances"
X = normalised_df.drop(columns=[target])
y = normalised_df[target]
```

```python
# split data into training and test sets with a 70:30 ratio and a random state of 42 (for reproducibility)

x_train, x_test, y_train, y_test = train_test_split(X, y,test_size=0.3, random_state=42)
```

### **Training the Algorithm**
We have split our data into training and testing sets, and now is finally the time to train our algorithm. 

```python
# instantiate LinearRegression model

linear_model = LinearRegression()

#fit the model to the training dataset
linear_model.fit(x_train, y_train)

#obtain predictions
predicted_values = linear_model.predict(x_test)
```

```python
# Mean Absolute Error (MAE)

mae = mean_absolute_error(y_test, predicted_values)
print('Mean Absolute Error:', round(mae, 2)) 
```

```python
# Residual Sum of Squares (RSS)

rss = np.sum(np.square(y_test - predicted_values))
print('Residual Sum of Squares:', round(rss, 2)) 
```

```python
# Root Mean Square Error (RMSE)

rmse = np.sqrt(mean_squared_error(y_test, predicted_values))
print('Root Mean Square Error:', round(rmse, 3)) 
```

```python
# R-Squared i.e. coefficient of determination

r2_score = r2_score(y_test, predicted_values)
round(r2_score, 2)
print('R-squared:', round(r2_score, 2)) 
```

Obtain the feature weights from your linear model above. Which features have the lowest and highest weights respectively?

```python
# obtain feature weights for linear regression model

coefficient = linear_model.coef_

feature_weight = pd.DataFrame(coefficient, X.columns, columns=["Coefficient"]).sort_values("Coefficient")
feature_weight
```

Train a ridge regression model with an alpha value of 0.4. 

Is there any change to the root mean squared error (RMSE) when evaluated on the test set?

```python
# instantiate ridge model

ridge_reg = Ridge(alpha=0.4)
ridge_reg.fit(x_train, y_train)

#obtain predictions
predicted_values = ridge_reg.predict(x_test)

# Root Mean Square Error (RMSE)

rmse = np.sqrt(mean_squared_error(y_test, predicted_values))
print('Root Mean Square Error:', round(rmse, 3)) 
```

Answer: There was no change for the RMSE of Linear Regression model and the Ridge Regression model.

Both models have RMSE value of 0.088

Train a lasso regression model with an alpha value of 0.001 and obtain the new feature weights with it. 

How many of the features have non-zero feature weights?

```python
# instantiate lasso model

lasso_reg = Lasso(alpha=0.001)
lasso_reg.fit(x_train, y_train)
```

```python
# obtain feature weights for lasso regression
coefficient = lasso_reg.coef_

feature_weight = pd.DataFrame(coefficient, X.columns, columns=["Coefficient"]).sort_values(by="Coefficient", ascending=False)
feature_weight
```

Answer: 4 features have non-zero feature weights

```python
#obtain predictions for lasso regression
predicted_values = lasso_reg.predict(x_test)

# Root Mean Square Error (RMSE)
rmse = np.sqrt(mean_squared_error(y_test, predicted_values))

print('Root Mean Square Error:', round(rmse, 3)) 
```

```python
#comparing the effects of regularisation

def get_weights_df(model, feat, col_name):
#this function returns the weight of every feature
    weights = pd.Series(model.coef_, feat.columns).sort_values()
    weights_df = pd.DataFrame(weights).reset_index()
    weights_df.columns = ['Features', col_name]
    weights_df[col_name].round(3)
    return weights_df

linear_model_weights = get_weights_df(linear_model, x_train, 'Linear_Model_Weight')
ridge_weights_df = get_weights_df(ridge_reg, x_train, 'Ridge_Weight')
lasso_weights_df = get_weights_df(lasso_reg, x_train, 'Lasso_weight')

final_weights = pd.merge(linear_model_weights, ridge_weights_df, on='Features')
final_weights = pd.merge(final_weights, lasso_weights_df, on='Features')

final_weights
```
